Podemos ver que de enero a mayo los datos tienen formato int64, pero para el resto del año están guardados en un formato float64. Para homogenizar los datos, y así evitar problemas futuros se realiza lo siguiente:

````python
lista = ['Enero','Febrero','Marzo','Abril','Mayo']
for i in lista:
  df[i]=df[[i]].astype(float)

# para verificar este cambio, una de las maneras es consultar de nuevo con
# df.info()
````

Para realizar una consulta de cuántos elementos hay en cada tipo de delito respecto al bien jurídico afectado:


````python
df_bja = df['Bien jurídico afectado'].value_counts() # se crea un nuevo marco de datos para contabilizar esto.
df_bja
````

Se puede realizar un conteo idéntico poara el Tipo de delito, Subtipo de delito y Modalidad con:


```python
df_bja = df['Tipo de delito'].value_counts()

df_bja = df['Subtipo de delito'].value_counts()

df_bja = df['Modalidad'].value_counts()
```
Al realizar estos calculos es notable que la cantidad de clasificaciónes en Tipo de delito, Subtipo de delito y Modalidad, son muchas y además son una particularidad del delito. Su estudio esclarece sólo el tipo de delito. Esto puede ser útil más a delante o no, dependiendo del tipo de pregunta que se busque responder. Por el momento nos será suficiente con graficar la cantidad de delitos por bien jurídico afectado, debido a que esta es una de las categorías más importantes al momento de clasificar los tipos de delitos.

Se procede a graficar la cantidad de delitos por bien jurídico afectado.












````python
df_bja=df_bja.reset_index() # resetear los índice y ordenarlos numéricamente.
df_bja=pd.DataFrame(df_bja) # se crea un data frame con los datos filtrados por bien jurídico afectado
df_bja
````

````python
# ahora hay que cambiarle el nombre a esas columnas
df_bja.rename(columns={"index" :'Bien jurídico afectado',
                        'Bien jurídico afectado':'cantidad'},
               inplace=True)
df_bja
````

````python
import matplotlib.pyplot as plt # importamos esta librería para hacer el gáfico
````

````python
plt.plot(df_bja['Bien jurídico afectado'],df_bja['cantidad'])   
plt.xlabel('Bien jurídico afectado', size=16)
plt.ylabel('Incidencias', size=16)
plt.xticks(rotation=60)

plt.title("Crimenes cometidos por bien jurídico afectado", 
          fontdict={'family': 'serif', 
                    'color' : 'darkblue',
                    'weight': 'bold',
                    'size': 18})
plt.grid(True)
plt.show()
````

````python
from matplotlib import cm
from matplotlib import colors

lista_bja= []
for i in df_bja['Bien jurídico afectado']:
  lista_bja.append(i)


lista_cantidad = []
for i in df_bja['cantidad']:
  lista_cantidad.append(i)

normdata = colors.Normalize(min(lista_cantidad), max(lista_cantidad))
colormap = cm.get_cmap("Blues")
colores =colormap(normdata(lista_cantidad))

plt.pie(lista_cantidad, labels=lista_bja, autopct="%0.1f %%", colors=colores)
plt.axis("equal")
plt.show()
````

De lo anterior podemos ver que de todos los delitos cometidos en todos estos años en México,la mayorían han tenido que ver con el patrimonio (42.9%)
