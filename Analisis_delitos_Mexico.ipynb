Se procede a graficar la cantidad de delitos por bien jur칤dico afectado.












````python
df_bja=df_bja.reset_index() # resetear los 칤ndice y ordenarlos num칠ricamente.
df_bja=pd.DataFrame(df_bja) # se crea un data frame con los datos filtrados por bien jur칤dico afectado
df_bja
````

````python
# ahora hay que cambiarle el nombre a esas columnas
df_bja.rename(columns={"index" :'Bien jur칤dico afectado',
                        'Bien jur칤dico afectado':'cantidad'},
               inplace=True)
df_bja
````

````python
import matplotlib.pyplot as plt # importamos esta librer칤a para hacer el g치fico
````

````python
plt.plot(df_bja['Bien jur칤dico afectado'],df_bja['cantidad'])   
plt.xlabel('Bien jur칤dico afectado', size=16)
plt.ylabel('Incidencias', size=16)
plt.xticks(rotation=60)

plt.title("Crimenes cometidos por bien jur칤dico afectado", 
          fontdict={'family': 'serif', 
                    'color' : 'darkblue',
                    'weight': 'bold',
                    'size': 18})
plt.grid(True)
plt.show()
````

````python
from matplotlib import cm
from matplotlib import colors

lista_bja= []
for i in df_bja['Bien jur칤dico afectado']:
  lista_bja.append(i)


lista_cantidad = []
for i in df_bja['cantidad']:
  lista_cantidad.append(i)

normdata = colors.Normalize(min(lista_cantidad), max(lista_cantidad))
colormap = cm.get_cmap("Blues")
colores =colormap(normdata(lista_cantidad))

plt.pie(lista_cantidad, labels=lista_bja, autopct="%0.1f %%", colors=colores)
plt.axis("equal")
plt.show()
````

De lo anterior podemos ver que de todos los delitos cometidos en todos estos a침os en M칠xico,la mayor칤an han tenido que ver con el patrimonio (42.9%)
